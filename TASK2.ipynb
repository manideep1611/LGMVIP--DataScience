```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import pandas as pd 
```

```python
#reading the image
img = cv2.imread("/content/drive/MyDrive/2-n.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
#converting to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
#inverting the image
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(10,10))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
#smoothening the image for pencil sketch
img_blur = cv2.GaussianBlur(img_invert, (21, 21), 0)
plt.figure(figsize=(10,10))
plt.imshow(img_blur,cmap='gray')
plt.axis('off')
plt.title('Blurred Inverted image')
plt.show()
```

```python
#final image
inverted_blurred = 255 - img_blur
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=256.0)
plt.figure(figsize=(10,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('Pencil Sketch')
plt.show()
```
